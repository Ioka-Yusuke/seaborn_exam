```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = (
        pd.read_csv("/workspaces/seaborn_exam/data/input/Nile.csv")
        .assign(hue=lambda x: "a")
    )

df_2 = (
        pd.read_csv("/workspaces/seaborn_exam/data/input/Nile.csv")
        .assign(
            hue=lambda x: "b",
            value= lambda x: x["value"] + 100
        )
    )

df = (
        pd.concat([df_1, df_2])
        .assign(
            category=lambda x: (x['time'] > 1921).astype(int) + 1
        )
    )


df.head(5)
```

```python
sns.set_context("talk")
sns.relplot(data=df, x="time", y="value", hue="hue", col="category",
            hue_order=["a", "b"], palette=["tab:orange", "tab:blue"], 
            height=6, aspect=1, legend=False)
# tab:gray
plt.subplots_adjust(hspace=0.2, wspace=0.2)

plt.xlim(1871, 1971)
plt.ylim(400, 1600)

plt.xticks([1871, 1921, 1971], [1871, 1921, 1971])
plt.yticks([500, 1000, 1500], [500, 1000, 1500])

plt.xlabel("time")
plt.ylabel("value")

# appendix
# plt.xticks(rotation=30)
# plt.axvline(x=50, color="red", linestyle="--", linewidth=2)
# plt.axhline(y=1000, color="blue", linestyle="--", linewidth=2)
```
