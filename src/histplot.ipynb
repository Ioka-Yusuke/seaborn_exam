```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = (
        pd.read_csv("/workspaces/seaborn_exam/data/input/Nile.csv")
        .assign(hue=lambda x: "a")
    )

df_2 = (
        pd.read_csv("/workspaces/seaborn_exam/data/input/Nile.csv")
        .assign(
            hue=lambda x: "b",
            value= lambda x: x["value"] + 100
        )
    )

df = (
        pd.concat([df_1, df_2])
        .assign(
            category=lambda x: (x['time'] > 1921).astype(int) + 1
        )
    )


df.head(5)
```

```python
sns.set_context("talk")
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="value", hue="hue", bins=100,
            hue_order=["a", "b"], palette=["tab:orange", "tab:blue"], 
            legend=False)
# tab:gray
plt.xlim(0, 1600)
plt.ylim(0, 7)

plt.xticks([0, 800, 1600], [0, 800, 1600])
plt.yticks([0, 3, 6], [0, 3, 6])

plt.xlabel("value")
plt.ylabel("cnt")
```

```python
sns.set_context("talk")
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="time", y="value", hue="hue", bins=10,
            hue_order=["a", "b"], palette=["tab:orange", "tab:blue"], 
            legend=False)
# tab:gray
plt.xlim(1860, 1980)
plt.ylim(400, 1600)

plt.xticks([1860, 1910, 1980], [1860, 1910, 1980])
plt.yticks([400, 800, 1600], [400, 800, 1600])

plt.xlabel("value")
plt.ylabel("cnt")
```
