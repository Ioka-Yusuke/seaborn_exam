```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = (
        pd.read_csv("/workspaces/seaborn_exam/data/input/Nile.csv")
        .assign(hue=lambda x: "a")
    )

df_2 = (
        pd.read_csv("/workspaces/seaborn_exam/data/input/Nile.csv")
        .assign(
            hue=lambda x: "b",
            value= lambda x: x["value"] + 100
        )
    )

df = (
        pd.concat([df_1, df_2])
        .assign(
            category=lambda x: (x['time'] > 1921).astype(int) + 1
        )
    )


df.head(5)
```

```python
sns.set_context("talk")
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="category", y="value", hue="hue",
            hue_order=["a", "b"], palette=["tab:orange", "tab:blue"], 
            legend=False)
# tab:gray
plt.axhline(100, c="tab:gray")

plt.ylim(0, 1200)

plt.xticks([0, 1], ["a", "b"])

plt.xlabel("time")
plt.ylabel("value")
```
