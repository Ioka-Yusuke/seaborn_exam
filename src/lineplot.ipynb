```python
import pandas as pd
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv('/workspaces/seaborn_exam/data/input/Nile.csv')
df_2 = df_1.copy()
df_2['value'] += 200
df_1['hue'] = 'a'
df_2['hue'] = 'b'
df = pd.concat([df_1, df_2])

df.head(5)
```

```python
sns.set_context('talk')
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='rownames', y='value', hue='hue',
             hue_order=['a', 'b'], palette=['tab:orange', 'tab:blue'],
             legend=False)
# tab:gray
plt.xlim(0, 100)
plt.ylim(400, 1600)
plt.xticks([0, 50, 100], [0, 50, 100])
plt.yticks([500, 1000, 1500], [500, 1000, 1500])
plt.xlabel('value')
plt.ylabel('time')
# plt.xticks(rotation=30)
# plt.axvline(x=50, color='red', linestyle='--', linewidth=2)
# plt.axhline(y=1000, color='blue', linestyle='--', linewidth=2)
```
